```python
import numpy as np
from scipy.stats import gamma
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('ggplot')
```

$Gamma(\theta |k, \lambda) \propto \theta^{k-1} \exp(-\lambda \theta)$

$\frac{dp}{dt}=(k-1)\frac{1}{\theta}-\lambda$

```python
eps = 1e-2
lam = 1
k = 11
NMCS = 10000
```

$H=\frac{1}{2} p^2 -(k-1)\log\theta +\lambda \theta$

$\frac{d\theta}{dt}=p$

$\frac{dp}{dt}=(k-1)\frac{1}{\theta}-\lambda$

```python
f = lambda x, lam,k :(k-1)/x-lam
```

```python
x,y = np.linspace(0,20,100),np.linspace(-3,3,100)
X,Y=np.meshgrid(x,y)
CS=plt.contour(X,Y,.5 * Y**2-(k-1)*np.log(X)+lam*X,levels=[-12,-10,-5,0])
plt.clabel(CS,inline=1,fontsize=10)
```

```python
def euler(q,p,eps,lam,k,f):
    q_new =q+eps*p
    p_new =p+eps*f(q,lam,k)
    q,p = q_new,p_new
    return q,p
```

```python
q,p =4.0, 0.0
euler_arr=np.zeros([NMCS,2])
euler_arr[0,0]=q
euler_arr[0,1]=p
for i in range(1,NMCS):
    q,p =euler(q,p,eps,lam,k,f)
    euler_arr[i,0]=q
    euler_arr[i,1]=p
df_lf=pd.DataFrame(euler_arr,columns=['q','p'])
```

```python
plt.plot(df_lf['q'],df_lf['p'],label='Euler')
CS=plt.contour(X,Y,.5 * Y**2-(k-1)*np.log(X)+lam*X,levels=[-12,-10,-5,0])
plt.clabel(CS,inline=1,fontsize=10)
```

```python
def leap_flog(q,p,eps,lam,k,f):
    p_new =  p+.5*eps *f(q,lam,k)
    q_new = q+eps*p_new
    p_new = p_new+.5*eps*f(q_new,lam,k)
    q,p = q_new,p_new
    return q,p
```

```python
q,p =4.0, 0.0
lf_arr=np.zeros([NMCS,2])
lf_arr[0,0]=q
lf_arr[0,1]=p
for i in range(1,NMCS):
    q,p =leap_flog(q,p,eps,lam,k,f)
    lf_arr[i,0]=q
    lf_arr[i,1]=p
df_lf=pd.DataFrame(lf_arr,columns=['q','p'])
```

```python
plt.plot(df_lf['q'],df_lf['p'],label='Euler')
CS=plt.contour(X,Y,.5 * Y**2-(k-1)*np.log(X)+lam*X,levels=[-12,-10,-5,0])
plt.clabel(CS,inline=1,fontsize=10)
```
